```python
#json - nosql

# importing libraries

import time
import json

# initialized a variable
Sale=''

#loading the inventory record from JSON to string 
fd=open('Records.json','r')
js=fd.read()
fd.close()

# converting String to json/dictionaries
record=json.loads(js)

#Displaying the menu
print("---------------------------Menu---------------------------\n")

print("Product ID | Product  \t| Price | Quantity")
for key in record.keys():
    print(key,":    ", record[key]['Name'],"\t|",record[key]['Price'],"\t|",record[key]['Qty'])
    
print("----------------------------------------------------------\n") 

#Taking user input about the product/purchase and user details
ui_name=str(input("Enter your name             :"))
ui_mail=str(input("Enter your e-mail ID,       :"))
ui_ph=str(input("Enter your phone number       :"))
ui_id=str(input("Enter the Product ID          :"))
ui_qty=int(input("Enter the Quantity           :"))


Total_amount=record[ui_id]['Price']*ui_qty


# if we are having equal or more quantity than user wants
if record[ui_id]['Qty']>=ui_qty:
    
    #Display purchase details
    print("----------------------------------------------------------\n")
    print("                           Bill                           \n")
    
    print("Name       :",record[ui_id]['Name'])
    print("Price      :",record[ui_id]['Price'])
    print("Quantity   :",ui_qty)
    print("Total amt  :",Total_amount)
    print("----------------------------------------------------------\n")
    
    
    #Generating the sales structure in CSV
    Sale= ui_name+","+ui_mail+","+ ui_ph+","+ui_id+","+record[ui_id]['Name']+","+str(ui_qty)+","+str(record[ui_id]['Price'])+","+str(Total_amount)+","+time.ctime()+'\n'

    
    
    if Total_amount >= 5000:
        # Apply a 10% discount
        discount=10
        dis_amt=Total_amount*(discount/100)
        final_price=Total_amount-dis_amt
        print("Original Price             :",Total_amount)
        print("Discount Amount (10% off)  :",dis_amt)
        print("----------------------------------------------------------\n")
        print("Final Price:", final_price)
        print("----------------------------------------------------------\n")
    else:
        print("Customer does not qualify for a discount")
    
    #updating  records Dictionary 
    record[ui_id]['Qty']=record[ui_id]['Qty']-ui_qty
    
    
#if we are having zero quantity
elif record[ui_id]['Qty']==0:
    print("Out of Stock!!!")
    
#if we are not having enough quantity     
else:
    
    #Displaying info about the available qty
    print("Sorry, we're not having enough quantity in our Inventory")
    print("We are only having", record[ui_id]['Qty'],"Quantity.")
    print("----------------------------------------------------------\n") 
    ch=str(input("Press Y/y to Purchase or N/n to exit: "))
    
    
    
    #if user wants to purchase the whole qty
    if ch=="Y" or ch=='y':
    
    #display purchase details
        print("----------------------------------------------------------\n")
        print("                           Bill                           \n")
        print("Name      :",record[ui_id]['Name'])
        print("Price     :",record[ui_id]['Price'])
        print("Quantity  :",record[ui_id]['Qty'])
        print("Total amt :",(int(record[ui_id]['Qty'])* int(record[ui_id]['Price'])))
        print("----------------------------------------------------------\n")
        
        #Generating the sales structure in CSV
        Sale= ui_name+","+ui_mail+","+ ui_ph+","+ui_id+","+record[ui_id]['Name']+","+str(record[ui_id]['Qty'])+","+str(record[ui_id]['Price'])+","+str(Total_amount)+","+time.ctime()+'\n'
        
        
        if Total_amount >= 5000:
            # Apply a 10% discount
            discount=10
            dis_amt=(int(record[ui_id]['Qty'])* int(record[ui_id]['Price']))*(discount/100)
            final_price=(int(record[ui_id]['Qty'])* int(record[ui_id]['Price']))-dis_amt
            print("Original Price             :",(int(record[ui_id]['Qty'])* int(record[ui_id]['Price'])))
            print("Discount Amount (10% off)  :",dis_amt)
            print("----------------------------------------------------------\n")
            print("Final Price:", final_price)
        else:
            print("Customer does not qualify for a discount")
        
        #update records Dictionaries
        record[ui_id]['Qty']=0
    
    #if user press any other key than y 
    else:
        print("Exited")
         

# converting Dictionary to String
js=json.dumps(record)

# saving records in JSON file
fd= open("Records.json",'w')
fd.write(js)
fd.close()

#saving records in Sales file
fd=open('Sales.txt','a')
fd.write(Sale)
fd.close()

#Thank you message
print("Thank You!!")
print("----------------------------------------------------------\n")
```
